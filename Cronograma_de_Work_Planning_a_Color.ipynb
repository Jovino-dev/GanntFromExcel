```
# prompt: leer un fichero excel subido por el usuario y leer el contenido de la primera hoja en un dataframe

from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # Check if the uploaded file is an Excel file
  if fn.endswith('.xlsx') or fn.endswith('.xls'):
    df = pd.read_excel(io.BytesIO(uploaded[fn]), sheet_name=0)  # Read the first sheet

    # Now you can work with the DataFrame 'df' as needed
  else:
    print("Uploaded file is not an Excel file.")
```

```
import altair as alt

# Create a new dataframe `df_gantt` with the columns 'Phase', 'Activity', 'Start', 'End', and 'Duration'
df_gantt = df.copy()[['Phase', 'Activity', 'Totals']]
df_gantt.rename(columns={'Totals': 'Duration'}, inplace=True)
#df_gantt = df_gantt[df_gantt['Activity'] != 'Total']
df_gantt = df_gantt[df_gantt['Phase'] != 'Total']

# Then proceed with the rest of your code
df_gantt = df_gantt[['Phase', 'Activity', 'Duration']] #Select the columns you need for the Gantt chart

# Forward fill the 'Phase' column to fill in the missing values
df_gantt['Phase'] = df_gantt['Phase'].ffill()

# Group by 'Phase' and calculate cumulative sum of 'Duration' to get 'End'
df_gantt['End'] = df_gantt['Duration'].cumsum()

# Desplaza la columna 'End' para obtener 'Start'
df_gantt['Start'] = df_gantt['End'].shift(1).fillna(0)
df_gantt = df_gantt[df_gantt['Activity'] != 'Total']

# Create the Gantt chart
chart = alt.Chart(df_gantt).mark_bar().encode(
    x=alt.X('Start:Q', title='Hours'),
    x2='End:Q',
    y=alt.Y('Activity:N', axis=alt.Axis(title='Activity')),
    color=alt.Color('Phase:N', legend=alt.Legend(title='Phase')),
    tooltip=['Phase', 'Activity', 'Start', 'End', 'Duration']
).properties(
    title='Project Work Planning Gantt Chart',
        width=1366,  # Set the width of the chart
    height=768  # Set the height of the chart
).interactive()
chart.show()
chart.save('gantt_chart.json')
# Save the chart as a JSON file
```
